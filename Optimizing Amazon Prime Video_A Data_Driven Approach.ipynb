## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
````

````python
# Handling missing values
credits_df.fillna("Unknown", inplace=True)
titles_df.fillna("Unknown", inplace=True)
````

````python
# Convert data types if necessary
titles_df['release_year'] = titles_df['release_year'].astype(int)
titles_df['runtime'] = titles_df['runtime'].astype(int, errors='ignore')
````

````python
# Normalize column formats
titles_df['genres'] = titles_df['genres'].apply(lambda x: eval(x) if isinstance(x, str) else [])
titles_df['production_countries'] = titles_df['production_countries'].apply(lambda x: eval(x) if isinstance(x, str) else [])
````

````python
# Drop irrelevant columns in credits_df
credits_df.drop(columns=['character'], inplace=True, errors='ignore')
````

````python
# Drop 'seasons' column in titles_df
titles_df.drop(columns=['seasons'], inplace=True, errors='ignore')
````

````python
# Replace 'Unknown' with NaN for numerical conversions
titles_df.replace('Unknown', np.nan, inplace=True)
````

````python
# Convert numeric columns properly, handling errors
titles_df.replace("Unknown", np.nan, inplace=True)
numeric_cols = ['runtime', 'imdb_score', 'imdb_votes', 'tmdb_popularity', 'tmdb_score']
for col in numeric_cols:
    titles_df[col] = pd.to_numeric(titles_df[col], errors='coerce')
````

````python
# Drop rows with any remaining NaN values in numeric columns before plotting
titles_df.dropna(subset=numeric_cols, inplace=True)
````

````python
# Ensure genres and production_countries are properly parsed
titles_df['genres'] = titles_df['genres'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else [])
titles_df['production_countries'] = titles_df['production_countries'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else [])
````

````python
# Handling missing values in titles_df
titles_df['age_certification'].fillna(titles_df['age_certification'].mode()[0], inplace=True)
titles_df['runtime'].fillna(titles_df['runtime'].median(), inplace=True)
titles_df['imdb_score'].fillna(titles_df['imdb_score'].mean(), inplace=True)
titles_df['imdb_votes'].fillna(titles_df['imdb_votes'].median(), inplace=True)
titles_df['tmdb_popularity'].fillna(titles_df['tmdb_popularity'].median(), inplace=True)
titles_df['tmdb_score'].fillna(titles_df['tmdb_score'].mean(), inplace=True)
````

````python
# Reset index after cleaning
titles_df.reset_index(drop=True, inplace=True)
credits_df.reset_index(drop=True, inplace=True)
````
